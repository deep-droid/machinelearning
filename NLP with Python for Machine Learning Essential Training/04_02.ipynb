# Feature Engineering: Feature Creation

### Read in text

```python
import pandas as pd

data = pd.read_csv("Ex_Files_NLP_Python_ML_EssT\\Exercise Files\\Ch04\\04_02\\Start\\SMSSpamCollection.tsv", sep='\t')
data.columns = ['label', 'body_text']
```

### Create feature for text message length

```python
data["body_len"] = data["body_text"].apply(lambda x: len(x) - x.count(" "))

data.head(100)
```

### Create feature for % of text that is punctuation

```python
import string

def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text) - text.count(" ")), 3) * 100

data["punct%"] = data["body_text"].apply(lambda x: count_punct(x))
data.head(200)
```

### Evaluate created features

```python
from matplotlib import pyplot
import numpy as np
%matplotlib inline
```

```python
bins = np.linspace(0, 200, 40)

pyplot.hist(data[data["label"] == "spam"]["body_len"], bins, alpha = 0.5, density=True, label="spam")
pyplot.hist(data[data["label"] == "ham"]["body_len"], bins="ham"), alpha = 0.5, density=True, label
pyplot.legend(loc="upper left")
pyplot.show()
```

```python
bins = np.linspace(0, 50, 40)

pyplot.hist(data[data["label"] == "spam"]["punct%"], bins, alpha = 0.5, density=True, label="spam")
pyplot.hist(data[data["label"] == "ham"]["punct%"], bins, alpha = 0.5, density=True, label="ham")

pyplot.legend(loc="upper right")
pyplot.show()
```
